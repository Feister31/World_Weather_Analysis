# Challenge Part 1 Get the weather description and amount of precipitation for each city

```python
from citipy import citipy
import numpy as np
import pandas as pd
```

```python
# Create a set of random latitudes and lontitudes combinations
lats = np.random.uniform(low =-90.000, high=90.000, size = 1500)
lngs = np.random.uniform(low =-180.000, high=180.000, size = 1500)
lat_lngs =zip(lats,lngs)
```

```python
# Add the latitudes and longtitudes to a list
coordinates = list(lat_lngs)
```

```python
# Create a list for holding the cities
cities = []

# identify nearest city for each latitude and longtitude combination 

for coordinate in coordinates:
    city = citipy.nearest_city(coordinate[0], coordinate[1]).city_name
    
    #if the city the unique then add it to cities list
    if city not in cities:
        cities.append(city)
len(cities)
```

```python
import requests

from datetime import datetime

from config import weather_api_key
```

```python
url = f"http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID={weather_api_key}"
print(url)
```

```python
# List of city data
city_data = []
# Print to logger
print("Beginning Data Retrivial")
print("-" * 25)

# Create counter
record_count = 1 
set_count = 1
for index, city in enumerate(cities):
    
    #Group cities in sets of 50 for logging puropose 
    if(index % 50 == 0 and index >= 50):
        set_count += 1
        record_count = 1
    #Create endpoint URL with each city
    city_url = f'{url}&q={city.replace(" ","+")}'
    print(f"Processing Record {record_count} of Set {set_count} | {city}")
    
    record_count += 1 
    
    try:
       # Parse the JSON and retrieve data.
        city_weather = requests.get(city_url).json()
    #print(city_url)
        # Parse out the needed data.
        city_lat = city_weather["coord"]["lat"]
        city_lng = city_weather["coord"]["lon"]
        city_max_temp = city_weather["main"]["temp_max"]
        city_humidity = city_weather["main"]["humidity"]
        city_clouds = city_weather["clouds"]["all"]
        city_wind = city_weather["wind"]["speed"]
        city_country = city_weather["sys"]["country"]
        city_weather_description = city_weather["weather"][0]["description"]
        city_rain_inches = 0
        city_snow_inches = 0
        # Convert the date to ISO standard.
        
        city_date = datetime.utcfromtimestamp(city_weather["dt"]).strftime('%Y-%m-%d %H:%M:%S')
        # Append the city information into city_data list.
        city_data.append({"City": city.title(),
                          "Lat": city_lat,
                          "Lng": city_lng,
                          "Max Temp": city_max_temp,
                          "Rain (inches)": city_rain_inches,
                          "Snow (inches)": city_snow_inches,
                          "Current Description": city_weather_description,
                          "Humidity": city_humidity,
                          "Cloudiness": city_clouds,
                          "Wind Speed": city_wind,
                          "Country": city_country,
                          "Date": city_date

        })
    except:
        print("City not found. Skipping....")
        pass
#indicate that the Data loading is complete
print("-" * 20)
print("Data Retrieval is complete")
print("-" * 20)
```
